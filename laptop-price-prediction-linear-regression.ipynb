```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/laptop-price-prediction-cleaned-dataset/laptop_data_cleaned.csv")
df.head()
```

```python
# Shape of the dataset
df.shape
```

```python
df.info()
```

```python
df=pd.get_dummies(df, columns=["Company", "TypeName", "Cpu_brand", "Gpu_brand", "Os"])
```

```python
df.corr()
```

## **Model Building**

```python
#Splitting input and output parameter
df = df[[col for col in df.columns if col != 'Price'] + ['Price']]
```

```python
X = df.iloc[:,:43]
y = df.iloc[:,43:]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state=100)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn import linear_model #Parametric Model
model = linear_model.LinearRegression()
```

```python
#Training(Contains the intelligence for estimating the values of 'm'and 'c', thus having the solved equation)
model.fit(X_train, y_train)
```

```python
# The coefficients
print('Coefficients', model.coef_)
print('Intercept', model.intercept_)
```

```python
#Testing
y_pred = model.predict(X_test)
y_pred_tr = model.predict(X_train)
```

```python
y_pred
```

```python
#Evaluation
#Testing accuracy
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_test, y_pred)) #Coefficient of Determination
print(mean_squared_error(y_test,y_pred))
```

```python
#Training accuracy
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_train, y_pred_tr)) #Coefficient of Determination
print(mean_squared_error(y_train, y_pred_tr)) #MSE
```
